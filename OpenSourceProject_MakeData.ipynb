```python
#강사모
text = dict()

for i in range(1000) :
    plus = "%d" %(i)
    url = 'https://cafe.naver.com/ArticleList.nhn?search.clubid=10258021&search.menuid=318&userDisplay=50&search.boardtype=L&search.specialmenutype=&search.totalCount=501&search.page='+plus
    req = requests.get(url)
    html = req.text;
    headers = req.headers;
    status = req.status_code;
    is_ok = req.ok
    soup = BeautifulSoup(html, 'html.parser')


    nw_list = soup.select('tr > td')
    for nw in nw_list:
        titles = nw.select('a.article')
        for title in titles:
            text_ap = title.get_text().strip().split()
            for text_new in text_ap :
                text_new = text_new.strip('.?,!)(@#$%^&*)_+|\=-/[]~')
            
                text_new = text_new.split('/')
            
                if type(text_new) == type(list()):
                    for text_new_one in text_new :
                        if(len(text_new_one) < 2):
                            continue
                    
                        if(text_new_one in delete_text):
                            continue
                    
                        if text_new_one not in text :
                            text[text_new_one] = 1
                        else : 
                            text[text_new_one] += 1
                else :
                    if(len(text_new_one) < 2):
                        continue
                    if(text_new_one in delete_text):
                        continue
                    if text_new not in text :
                        text[text_new] = 1
                    else : 
                        text[text_new] += 1
```

```python
text_sort = sorted(text.items(),key = lambda x: x[1], reverse = True)
text_sort
```

```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
data = df 
train_df = data.iloc[:30, : ]
test_df = data.iloc[30:, :]

X = np.array(train_df['강사모']).reshape(-1,1)
y = train_df['러브캣']
lr = LinearRegression()
lr.fit(X, y)
#print(lr.coef_)

w = lr.coef_[0]

plt.scatter(X, y, s=1)
plt.plot(X, w*X, c='red')

```

위의 그래프는 '강사모'를 이용하여 '러브캣'의 개수를 예측한 그래프이다.
빨간색 선은 강사모 1000 : 러브캣 800을 가르키고 있다.
이는 강사모에 있는 검색어들은 러브캣에서도 사용 할 수 있다는것을 알 수 있다.

```python
test_X = np.array(test_df['강사모']).reshape(-1,1)
pred_y = lr.predict(test_X)

plt.scatter(test_X, pred_y)
plt.plot(test_X, w*test_X, c='red')

test_y = test_df['러브캣']

plt.scatter(test_X, test_y, c='purple', s=1)
```

하지만 test data 값을 넣어보면 위의 train data값과는다른 추세선이 그려진다.
이를 통해 알 수 있는 것은 하나의 사이트에서 다른 사이트로의 연관성이 떨어진다는 것을 알 수 있다.

```python
from sklearn.metrics import mean_squared_error
test_loss = mean_squared_error(test_y, pred_y)

train_y = lr.predict(X)
train_loss = mean_squared_error(y, train_y)

print(test_loss, train_loss)
plt.scatter(test_y, pred_y)
plt.plot([0,np.max(test_y)],[0,np.max(test_y)], color='red')
plt.show()

from sklearn.metrics import r2_score
r2 = r2_score(test_y, pred_y)
print(r2)
```

test에서의 예측값이 실제 데이터 값과 많이 다르다는 것을 알 수 있다. 서로 다른 두 사이트간에 단어에 대한 연관성은 없다는 것을 알 수 있다.
